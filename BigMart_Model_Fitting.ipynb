```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score
```

```python
df=pd.read_csv('Preprocessed_Data.csv')
df.head()
```

```python
x=df.drop(['Item_Outlet_Sales','Unnamed: 0'],axis='columns')
y=df['Item_Outlet_Sales']
```

```python
x
```

```python
y
```

### **Model Fitting**

#### **1. Linear Regression**

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
score_lr=cross_val_score(lr,X=x,y=y,cv=5)
```

```python
print(f"Score for Linear Regression Model : {np.mean(score_lr)}")
```

#### **2. Lasso Regression**

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso_parameter={
    'alpha':[0.0001,0.0005,0.0009,0.01,0.05,0.09,0.1,0.5,0.9]
}
lasso_reg=GridSearchCV(Lasso(),param_grid=lasso_parameter,cv=5,n_jobs=-1,verbose=1,return_train_score=False)
lasso_reg.fit(x,y)
```

```python
print(f'Best score : {lasso_reg.best_score_}')
print(f'Best Parameter : {lasso_reg.best_params_}')
```

#### **3. Ridge Regression**

```python
from sklearn.linear_model import Ridge
ridge_parameters={
    'alpha':[0.0001,0.0005,0.0009,0.01,0.05,0.09,0.1,0.5,0.9]
}
ridge_reg=GridSearchCV(Ridge(),param_grid=ridge_parameters,cv=5,n_jobs=-1,verbose=1,return_train_score=False)
ridge_reg.fit(x,y)
```

```python
print(f'Best Score : {ridge_reg.best_score_}')
print(f'Best Parameter : {ridge_reg.best_params_}')
```

#### **4. Random Forest Regressor**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV

rf_parameter={
    'n_estimators':[100,150,200,250],
    'max_depth': [5,6,7,8],
    'min_samples_split': [8,9,10,11],
    'min_samples_leaf': [20,22,24,25]
}
random_forest=RandomizedSearchCV(RandomForestRegressor(),param_distributions=rf_parameter,n_iter=100,verbose=2,return_train_score=False)
random_forest.fit(x,y)
```

```python
print(f'Best Score : {random_forest.best_score_}')
print(f'Best Parameter : {random_forest.best_params_}')
```
